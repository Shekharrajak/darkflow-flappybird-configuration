```python
import cv2
from darkflow.net.build import TFNet
import matplotlib.pyplot as plt

#To show this as svg images
%config InlineBackend.figure_format = 'svg'
```

```python
options = {
    'model': 'cfg/tiny-yolo-voc-1c.cfg',
    'load': 1000,                            
    'threshold': 0.1,                       
                            
}

tfnet = TFNet(options)
```

```python
# read the color image and covert to RGB

img = cv2.imread('new_model_data/images/3.jpg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# use YOLO to predict the image
results = tfnet.return_predict(img)

img.shape
```

```python
import numpy as np

colors = [tuple(255 * np.random.rand(3)) for i in range(10)]
# results = tfnet.return_predict(img)

for color, result in zip(colors, results):
    tl = (result['topleft']['x'], result['topleft']['y'])
    br = (result['bottomright']['x'], result['bottomright']['y'])
    label = result['label']
    frame = cv2.rectangle(img, tl, br, color, 7)
#     frame = cv2.putText(img, label, tl, cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 0), 2)
# cv2.imshow('frame', img)
plt.imshow(img)
plt.show()
```

```python
colors = [tuple(255 * np.random.rand(3)) for i in range(20)]
```

```python
results = sorted(results, key= lambda x: x['bottomright']['x'])
results
```

```python
flappybird_index = 0
for i, result in enumerate(results):
    tl = (result['topleft']['x'], result['topleft']['y'])
    br = (result['bottomright']['x'], result['bottomright']['y'])
    label = result['label']
    frame = cv2.rectangle(img, tl, br, colors[i], 7)
    if label =='flappybird':
        flappybird_index = i

# Flappy bird (x,y)
flappybird_xy = (results[flappybird_index]['bottomright']['x'],results[flappybird_index]['bottomright']['y'])
next_pipe_xy = (results[flappybird_index+1]['topleft']['x'],results[flappybird_index+1]['topleft']['y'])

print('Flappy bird is at: {}'. format(flappybird_xy))
print('Flappy bird - next green pipe is at: {}'. format(next_pipe_xy))

plt.imshow(img)
plt.show()
```
